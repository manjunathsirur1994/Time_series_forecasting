```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tbats import TBATS
import yfinance as yf
```

```python
ticker = 'LLOYDSME.NS'

df = yf.download(ticker, start='2020-01-01', multi_level_index=False)
df
```

```python
df = df[['Close']]
df = df.reset_index()
df
```

```python
df.shape
```

```python
df_train = df.head(325)
df_test = df.tail(1)
```

```python
df_train
```

```python
df_test
```

```python
model = TBATS(seasonal_periods=(7, 30, 365))
model = model.fit(df_train['Close'])
```

```python
y_pred = model.forecast(steps=1)
y_pred
```

```python
df_test
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(df_test['Close'], y_pred)
```

```python
import math

math.sqrt(mean_squared_error(df_test['Close'], y_pred))
```
